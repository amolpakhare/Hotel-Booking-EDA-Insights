Summary of Columns:-
1. **hotel**: Indicates the type of hotel — either *City Hotel* or *Resort Hotel*.

2. **is\_canceled**: Shows whether the booking was canceled (`1`) or not (`0`).

3. **lead\_time**: Number of days between the booking date and the actual arrival date.

4. **arrival\_date\_year**: The year in which the customer arrived at the hotel.

5. **arrival\_date\_month**: The month in which the customer arrived.

6. **arrival\_date\_week\_number**: Week number of the year when the customer arrived.

7. **arrival\_date\_day\_of\_month**: Day of the month when the customer arrived.

8. **stays\_in\_weekend\_nights**: Number of weekend nights (Saturday/Sunday) the customer stayed.

9. **stays\_in\_week\_nights**: Number of weekdays (Monday–Friday) the customer stayed.

10. **adults**: Number of adults included in the booking.

11. **children**: Number of children included in the booking.

12. **babies**: Number of babies included in the booking.

13. **meal**: Type of meal plan booked (e.g., BB, HB, FB, SC).

14. **country**: Country of origin of the customer.

15. **market\_segment**: How the customer made the booking (e.g., Direct, Corporate, Online TA, Offline TA).

16. **distribution\_channel**: Channel used to distribute the booking (e.g., Direct, TA/TO, GDS).

17. **is\_repeated\_guest**: Indicates if the customer is a returning guest (`1`) or not (`0`).

18. **previous\_cancellations**: Number of previous bookings the customer canceled.

19. **previous\_bookings\_not\_canceled**: Number of past bookings that were not canceled.

20. **reserved\_room\_type**: Code for the room type the customer initially reserved.

21. **assigned\_room\_type**: Code for the room type the customer was actually assigned.

22. **booking\_changes**: Number of times the booking was modified.

23. **deposit\_type**: Type of deposit the customer made (e.g., No Deposit, Non Refund, Refundable).

24. **agent**: ID of the travel agent or intermediary who made the booking.

25. **company**: ID of the company that made the booking (mainly for corporate bookings).

26. **days\_in\_waiting\_list**: Number of days the booking stayed on the waiting list before confirmation.

27. **customer\_type**: Category of the customer (e.g., Transient, Group, Contract, Transient-Party).

28. **adr** *(Average Daily Rate)*: The average amount the customer paid per day.

29. **required\_car\_parking\_spaces**: Number of car parking spaces requested by the customer.

30. **total\_of\_special\_requests**: Number of special requests made (e.g., high floor, crib, late checkout).

31. **reservation\_status**: Final status of the reservation (e.g., Check-Out, Canceled, No-Show).

32. **reservation\_status\_date**: Date when the reservation status was last updated.

# Data Preprocessing/ Data Preparation

```python
# Change the column datatype
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], dayfirst=True)
```

```python
#Replace the Column values and change datatype
df['is_repeated_guest'].replace({0:'No',1:'Yes'},inplace=True)
df['is_canceled'].replace({0:'No',1:'Yes'},inplace=True)
```

```python
continuous = ['lead_time', 'adr']

count = ['booking_changes', 'adults', 'children', 'babies', 
         'previous_cancellations', 'previous_bookings_not_canceled',
         'required_car_parking_spaces', 'total_of_special_requests',
         'stays_in_weekend_nights', 'stays_in_week_nights']

categorical = ['hotel', 'meal', 'market_segment', 'distribution_channel', 
               'reserved_room_type', 'assigned_room_type', 'deposit_type', 
               'reservation_status','country', 'customer_type','is_repeated_guest','is_canceled']

date_time = ['arrival_date_year', 'arrival_date_month', 'arrival_date_week_number', 
             'arrival_date_day_of_month', 'reservation_status_date']
```

```python
for col in categorical:
    print(col)
    print(df[col].unique())
    print('-' * 6)
```

```python
for col in count:
    print(col)
    print(df[col].unique())
    print('--' *6)
```

```python
for col in continuous:
    print(col)
    print(df[col].nunique())
    print('--' * 6)
```

```python
# Handling Missing Values 
df.drop(['company','agent'],axis=1,inplace=True)
df.dropna(inplace=True)
```
